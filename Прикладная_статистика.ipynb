01-03 Характеристики случайных величин

Математическое ожидание E(X) - среднее значение случайной величины

Для дискретной: взять каждое значение, которое оно может принять и умножить на вероятность
этих значений, а потом всё просуммировать.
E(X) = (сумма k=1 до n) k* P(X=k)

Для непрерывной: умножить каждое её значение на плотность распределения вероятности,
которое ему соответствует. Но т.к. это непрерывная величина, вместо суммы берем интеграл (аналог суммы)
E(X) = интеграл от -беск до +беск t* f(t)dt

Свойства математического ожидания
X, Y - случ величины
a - константа

1. E(a) = a; 	(42 всегда 42,
т.к. это не случ величина, а детерминированная)
2. E(X+Y) = E(X) + E(Y)
3. E(a*X)=a*E(X)
4. E(X*Y)=E(X)*E(Y), если независимы.
5. Мат ожидание случ величины - не случайно (это какая-то константа, которая описывает
то типичное среднее знач, которое эта случ величина принимает)
6. Если мы хотим понять, насколько в среднем наша случ величина отклоняется
от своего мат ожидания E(X - E(X)), то:
E(X - E(X))= E(X) - E(E(X))= E(X) - E(X) = 0

Чтобы понять, насколько наша случ величина отклоняется от мат ожидания,
то искать E(X-E(X)) бессмысленно, т.к. оно равно 0.
Для вычисления этого вводят понятие дисперсия.

Дисперсия - мера разброса случайной величиный вокруг её среднего.
